```python
#数据处理
def read_data(path):
    contents,labels = [],[]
    with open(path,'r',encoding='utf-8',errors='ignore') as fo:
        for line in fo:
            try:
                label,content = line.strip().split('\t')
                if content:
                    contents.append(content)
                    labels.append(label)
            except: pass
    fo.close()
    return contents,labels

def read_vocab():
    with open('./data/cnews.vocab.txt','r',encoding='utf-8',errors='ignore') as fo:
        words = [_.strip() for _ in fo.readlines()] 
    word_to_id = dict(zip(words,range(len(words))))
    fo.close()
    return words,word_to_id

def read_category():
    categories = ['体育', '财经', '房产', '家居', '教育', '科技', '时尚', '时政', '游戏', '娱乐']
    cat_to_id = dict(zip(categories, range(len(categories)))) 
    return categories,cat_to_id

def data_to_id(path,word_to_id,cat_to_id):
    contents,labels = read_data(path)
    contents_id,labels_id = [],[]
    for i in range(len(contents)):
        contents_id.append([word_to_id[x] for x in contents[i] if x in word_to_id])
        labels_id.append(cat_to_id[labels[i]])
    #使用keras提供的pad_sequences来将文本pad为固定长度
    contents_id = kr.preprocessing.sequence.pad_sequences(contents_id, 600)
    # 将标签转换为one-hot表示
    labels_id = kr.utils.to_categorical(labels_id, num_classes=len(cat_to_id))
    return contents_id,labels_id

words,word_to_id = read_vocab()
categories,cat_to_id = read_category()
x_train,y_train = data_to_id('./data/cnews.train.txt',word_to_id,cat_to_id)
x_test,y_test = data_to_id('./data/cnews.test.txt',word_to_id,cat_to_id)
x_val,y_val = data_to_id('./data/cnews.val.txt',word_to_id,cat_to_id)
```

```python
#定义超参
EPOCH = 10
INPUT_SIZE = 64
HIDDEN_SIZE = 128
NUM_LAYERS = 2
BATCH_SIZE = 128
LR = 0.001
```

```python
#数据分批
train_dataset = Data.TensorDataset(torch.LongTensor(x_train),torch.Tensor(y_train))
train_loader = Data.DataLoader(dataset=train_dataset,batch_size=BATCH_SIZE,shuffle=True)
test_dataset = Data.TensorDataset(torch.LongTensor(x_test),torch.Tensor(y_test))
test_loader = Data.DataLoader(dataset=test_dataset,batch_size=BATCH_SIZE,shuffle=True)
val_dataset = Data.TensorDataset(torch.LongTensor(x_val),torch.Tensor(y_val))
val_loader = Data.DataLoader(dataset=val_dataset,batch_size=BATCH_SIZE,shuffle=True)
```

```python
#定义模型
class TextRNN(nn.Module):
    def __init__(self,padsize):
        super(TextRNN,self).__init__()
        self.padsize = padsize
        self.embedding = nn.Embedding(5000,64)
        self.rnn = nn.GRU(input_size = INPUT_SIZE,hidden_size = HIDDEN_SIZE,num_layers = NUM_LAYERS,bidirectional = True)
        self.f1 = nn.Sequential(nn.Linear(256*self.padsize,128),nn.ReLU(),nn.Dropout(0.8))
        self.f2 = nn.Sequential(nn.Linear(128,10))
    def forward(self,x):
        x = self.embedding(x)
        x,_ = self.rnn(x)
        x = x.view(-1,256*self.padsize)
        x = self.f1(x)
        return self.f2(x)
model = TextRNN(600)
#print(model.padsize)
```

```python
#设置GPU
device = torch.device("cuda:0" if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
#设置criterion和optimizer
criterion = nn.MultiLabelSoftMarginLoss()
optimizer = optim.Adam(model.parameters(),lr=LR)
```

```python
#训练并保存
best_val_acc = 0
for epoch in range(EPOCH):
    model.train()
    for i,(x,y) in enumerate(train_loader):
        x = x.to(device)
        y = y.to(device)
        out = model(x)
        loss = criterion(out,y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        out = torch.softmax(out,dim=1)
        accuracy = np.mean((torch.argmax(out,1)==torch.argmax(y,1)).cpu().numpy())
        if i%100 == 99:
            print("epoch:{},step:{},loss:{},accuracy:{}".format(epoch+1,i+1,loss,accuracy))
    
    model.eval()
    epoch_val_acc = 0
    for i,(x,y) in enumerate(test_loader):
        x = x.to(device)
        y = y.to(device)
        out = model(x)
        step_val_acc = np.mean((torch.argmax(out,1)==torch.argmax(y,1)).cpu().numpy())
        epoch_val_acc += step_val_acc
        if step_val_acc>best_val_acc:
            torch.save(model.state_dict(),'./TextLen600.pt')
            print("New modelstate was changed! epoch:{},step:{},val_acc:{}".format(epoch,i,step_val_acc))
            best_val_acc = step_val_acc
    if best_val_acc>0.999:break
    epoch_val_acc = epoch_val_acc/(i+1)
```
